```python
data1 = data[['FIRE_YEAR', "NWCG_GENERAL_CAUSE", "STATE"]].groupby(["FIRE_YEAR", "NWCG_GENERAL_CAUSE", "STATE"]).value_counts().to_frame('Count').reset_index().rename({'index': 'value'})

data1.to_csv('C:\\Users\sammi\Desktop\Final Project Viz\\fire_cause.csv')
```

```python
# https://www.geeksforgeeks.org/pandas-group-by-multiple-columns/
# https://www.geeksforgeeks.org/python/pandas-groupby-count-the-occurrences-of-each-combination/


display(data[['FIRE_YEAR', "NWCG_GENERAL_CAUSE", "STATE"]].groupby(["FIRE_YEAR", "NWCG_GENERAL_CAUSE", "STATE"]).value_counts().to_frame('Count').reset_index().rename({'index': 'value'}))
```

```python
display(data[["FIRE_YEAR", "STATE", 'NWCG_REPORTING_AGENCY']].groupby(["FIRE_YEAR", "STATE", 'NWCG_REPORTING_AGENCY']).value_counts().to_frame('Count').reset_index().rename({'index': 'Value'}))
```

```python
# Reference: https://www.geeksforgeeks.org/r-data-analysis/how-to-name-the-column-when-using-valuecounts-function-in-pandas/

data2 = data[["FIRE_YEAR", "STATE", 'NWCG_REPORTING_AGENCY']].groupby(["FIRE_YEAR", "STATE", 'NWCG_REPORTING_AGENCY']).value_counts().to_frame('Count').reset_index().rename({'index': 'Value'})

data2.to_csv('C:\\Users\sammi\Desktop\Final Project Viz\\fire_agency.csv')
```
